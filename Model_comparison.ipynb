In this notebook, both models will be compared so the best can be determined by their maximum "mean Average Presicion" (mAP) score.

##First step, get CSVs
Both models have outputed a CSV file of their own with results and metrics, and analyzing the files, it was noted that both of them have the `mAP` metric, so it will be used to determine the best model for this dataset.

To parse the CSVs, the `pandas` library was used.

```python
# mount personal Google Drive
from google.colab import drive
drive.mount("/content/drive")
```

```python
# Load data
import pandas as pd
result_fast = pd.read_csv("/content/drive/MyDrive/Artificial Vision/project2/csv_files/results_faster-rcnn.csv")
result_yolo = pd.read_csv("/content/drive/MyDrive/Artificial Vision/project2/csv_files/results_yolov5.csv")
```

## Next, compare and determine
Using pandas, the dataframed CSVs were read and the `mAP` column was selected. Once selected the max value from both will be stored and both will be compared.

To improve the output display, a bar graph was plotted.

```python
import matplotlib.pyplot as plt

# Extract mAP scores
map_fast = max(result_fast['map_05'])
map_yolo = max(result_yolo['     metrics/mAP_0.5'])

# Print mAP scores
print(f"Faster_rcnn results = {map_fast}")
print(f"yolov5 results = {map_yolo}")

# Compare and print the best model
if map_fast > map_yolo:
  print(f"The model Faster R-CNN obtained a better mAP score: {round(map_fast*100, 2)}%")
else:
  print(f"The model YOLOv5 obtained a better mAP score: {round(map_yolo*100, 2)}%")

# Create a bar chart
models = ['Faster R-CNN', 'YOLOv5']
mAP_scores = [map_fast, map_yolo]

plt.bar(models, mAP_scores, color=['blue', 'purple'])
plt.xlabel('Object Detection Models')
plt.ylabel('mAP Score')
plt.title('Comparison of mAP Scores between Faster R-CNN and YOLOv5')
plt.show()
```
